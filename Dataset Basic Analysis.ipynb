```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
ds = pd.read_csv('./train.csv')
data = ds.values
ds.head(n=5)
```

```python
ds.describe()
```

```python
LABELS = {0:'T-Shirt', 1:'Trouser', 2:'Pullover', 3:'Dress', 4:'Coat', 5:'Sandal', \
          6:'Shirt', 7:'Sneaker', 8:'Bag', 9:'Ankle_Boot'}

images = []
plt.figure(0, figsize=(15,5))

for ix in range(10):
    ds2 = ds.drop(ds[ds.label!=ix].index)
    del ds2['label']
    image = np.asarray([ds2.mean(axis=0)]).T
    image = image.reshape(28,28)
    plt.subplot(2, 5, ix+1)
    plt.axis('off')
    plt.title(LABELS[ix])
    plt.imshow(image, cmap='gray')

plt.show()
```
